# Market Basket Analysis -

#### Market Basket Analysis is a technique which identifies the strength of association between pairs of products purchased together and identify patterns of occurrence. It helps us to do positioning of product at a store which generate more revenue and it helps our organization to grow. 


```python
# Generate data  --
```

```python
# manually create data  --

data = [['Shirt', 'Trousers', 'Belt', 'Socks', 'Boot', 'Watch'],
        ['T-shirt', 'Jacket', 'Jeans', 'Shorts', 'Sneakers', 'Socks'],
        ['Jacket', 'Gloves', 'T-shirt'],
        ['T-shirt', 'Jeans', 'Jacket'],
        ['T-shirt', 'Jeans', 'Sneakers','Socks', 'Jacket', 'Belt', 'Watch', 'Gloves'],
        ['Jeans', 'Sneakers', 'Belt', 'Socks']]
data
```

```python
# first we do  --

# transactiion_encoder = in which it converts data into true-false, like first it take all unique values from our dataset
# and put it all in a list then it check list wise data from this unique list, if those list item present in this unique
# list it make these item true and at absence place it gives false


from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()

newdata = encoder.fit(data).transform(data)
```

```python
newdata
```

```python
# it is important to convert it into dataframe otherwise we doesn't understand  --

import pandas as pd

dataset = pd.DataFrame(newdata, columns = encoder.columns_)

dataset
```

```python
# using Apriori and Association  --
```


For this we use two algorithm â€“ Apriori  &  Association  ->

Association Rule - The main goal of association rules is to identify relations between products or variables in a dataset.
The idea is to determine which products often come together.

Apriori Rule -  It is the most popular algorithm for mining association rules. It finds the most frequent combinations 
in a database and identifies association rules between the items, based on 3 important 


```python

#  ----
#importing apriori and association  --
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# dtermining the frequent factor -
frequent = apriori(dataset, use_colnames = True, min_support = 0.1, max_len = 4)

# filtering the factor on the basis of condition -
rules = association_rules(frequent, metric = 'lift', min_threshold = 1)

#Display the first 10 records  --
print(rules.head(15).sort_values(by = 'confidence', ascending = False))
```

```python
# list of product which were sold atleast 30% of time  --

apriori(dataset, use_colnames = True, min_support = 0.4)
```

```python

# it gives length of the combination of product -
frequent_itemsets = apriori(dataset, min_support = 0.4, use_colnames = True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x : len(x))
frequent_itemsets
```
